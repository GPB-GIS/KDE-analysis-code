```python
# Set the location of our voronoi clips to be extract
shapefiles = "../../data/working_files/ind_voronois.gdb"
```

```python
for i in layerlist[:5]:
    shp = gpd.read_file(shapefiles,layer=i)
    shp.crs = {'init' :'epsg:4326'} 
    shp.plot()
```

```python
# Create grid

# set the Grid height
squareWidth = 0.004
squareHeight = 0.004

# iterate through our layers
for i in layerlist:
    shp = gpd.read_file(shapefiles,layer=i)
    shp.crs = {'init' :'epsg:4326'} 
    #print (i)
    #print (shp_i)
    #shp.head()
   
    # addition of intersection with Nepal boundaries to remove unnecessary grids (speeds things up A LOT)
    # shouldnt be necessary as voronois are now pre-clipped (13/04/19)
    #shp_i = gpd.overlay(nepal_shp, shp, how='intersection')
 
    # get the boundings of our Voronoi 
    xmin,ymin,xmax,ymax = shp.total_bounds
    #print('1 ' + str(shp_i.total_bounds))
        
    xdiff = xmin - nxmin_p
    ydiff = nymax_p - ymax
    #print('2 ' + str(xdiff) + str(ydiff))
        
    gridWidth = xmax-xmin
    gridHeight = ymax-ymin
    
    # should we add a plus one here to offest the blanks??
        
    rows = int(np.ceil((ymax-ymin) /  squareHeight)) + 1
    cols = int(np.ceil((xmax-xmin) / squareWidth)) + 1
        
    XleftOrigin = nxmin_p + (int((xdiff/squareWidth))*squareWidth)
    XrightOrigin = nxmin_p + ((int((xdiff/squareWidth))*squareWidth) + squareWidth)
    
    YtopOrigin = nymax_p - ((int((ydiff/squareHeight)+1))*squareHeight)
    YbottomOrigin = nymax_p - ((int((ydiff/squareHeight)+1)*squareHeight) - squareHeight)
        
    #print ("Starting Points:", XleftOrigin, YtopOrigin)
        
    polygons = []
        
    for j in range(cols):
        Ytop = YtopOrigin
        Ybottom =YbottomOrigin
        for k in range(rows):
            polygons.append(Polygon([(XleftOrigin, Ytop), (XrightOrigin, Ytop), (XrightOrigin, Ybottom), (XleftOrigin, Ybottom)])) 
            Ytop = Ytop - squareHeight
            Ybottom = Ybottom - squareHeight
        XleftOrigin = XleftOrigin + squareWidth
        XrightOrigin = XrightOrigin + squareWidth
                
    grid = gpd.GeoDataFrame({'geometry':polygons})
    grid.crs = {'init' :'epsg:4326'}
        
    #print ("We've made a grid:", grid.total_bounds)
        
    voronoi_grid_clip = gpd.overlay(shp, grid, how='intersection')
    voronoi_grid_clip.crs = {'init' :'epsg:4326'}
    #voronoi_grid_clip.to_file("./working_files/voronoi_grid_clips/" + i + ".shp") 
    #print ("Voronoi grid clip created", voronoi_grid_clip.total_bounds)
        
    #Removes noise created by overlay function to leave only grid polygos, plus adds a potential ID field through index_righ
    left_df = voronoi_grid_clip
    right_df = grid
    gridclip_withids = gpd.sjoin(left_df, right_df, how='inner', op='within', lsuffix='left', rsuffix='right')

    gridclip_withids = gridclip_withids.rename(columns={'index_right': 'IID'})

    # Exports final grid clip to file
    #gridclip_withids.to_file("../../data/working_files/" + i + "_grid.shp")
    #print ("Grid clip exported")
    

    #  Conduct zonal stats on file
    zs_voronoi = zonal_stats(gridclip_withids, worldpop_raster_fp, nodata=-999.0, stats='sum', geojson_out=True)
    #print('here')

    # Export result as geojson
    result = {"type": "FeatureCollection","features": zs_voronoi}
    import geojson
    with open ("../../data/working_files/voronoi_grids/" + i + "_grid.geojson", 'w') as outfile:
        geojson.dump(result, outfile)
    
    voronoi_pop_shp = gpd.read_file("../../data/working_files/voronoi_grids/" + i + "_grid.geojson")
    voronoi_pop_shp = voronoi_pop_shp.rename(columns={'sum': 'Total_Pop'})
    voronoi_pop_sum = voronoi_pop_shp["Total_Pop"].sum()

    #print (voronoi_pop_sum)
    
    if isinstance(voronoi_pop_sum, float):
        
        voronoi_pop_shp['Pro_Pop'] = 0.0
        
        for m in range(len(voronoi_pop_shp['Total_Pop'])):
            voronoi_pop_shp.loc[m,'Pro_Pop'] = (voronoi_pop_shp.loc[m,'Total_Pop'])/voronoi_pop_sum 

        voronoi_pop_shp_edit =voronoi_pop_shp.drop(['id','Input_FID'], axis=1)
        voronoi_pop_shp_edit_order = voronoi_pop_shp_edit[['IID', 'geometry', 'Total_Pop', 'Pro_Pop']]

        voronoi_pop_shp_edit_order.to_file("../../data/working_files/voronoi_pop_grids/" + i + "_pop_v.shp")
        
        print (i + " processed!")
        
    else:
        voronoi_pop_shp.to_file("../../data/working_files/voronoi_pop_grids/" + i + "_FLAG_pop_v.shp")
        print (i + 'Flagged!')

print ("Processing Complete")
```

```python
# All flagged shapes checked and deemed too small to have been able to conduct zonal statistics. 
# Enter 1 in proportion column instead so all members are accounted for

path = "../../data/working_files/voronoi_pop_grids/"

for file in os.listdir(path):
    if file.endswith("FLAG_pop_v.shp"):
        voronoi = file[0:4]
        print (voronoi)
        shp_j = gpd.read_file(path + file)
        shp_j['Pro_Pop'] = 0.0
        
        for m in range(len(shp_j['IID'])):
            shp_j.loc[m,'Pro_Pop'] = 1
        
        shp_j_edit = shp_j.drop(['id','Input_FID'], axis=1)
        shp_j_edit_order = shp_j_edit[['IID', 'geometry', 'Pro_Pop']]
        shp_j_edit_order.to_file("../../data/working_files/voronoi_pop_grids/" + voronoi + "_pop_v.shp")
    
        print (file + " processed!")
```
