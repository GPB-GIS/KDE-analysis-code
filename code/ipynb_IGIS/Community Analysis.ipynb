```python
# Load in Libraries
import pandas as pd
import geopandas as gpd
import os
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# data path
data_path = "../../data/raw/ind_comm/"

# First set our empty list
comm_pop_datasets = []

# For each file within our processed folder, listed by the listfunction 
for file in os.listdir(data_path):
    # if the filename is equal to wildcard.shp, i.e. has a .shp extension
    if file.endswith('.csv'):
        # we use the .append() function to append the file to our crime_rate_datasets list
        comm_pop_datasets.append(file)

print(comm_pop_datasets)
```

```python
for comm in comm_pop_datasets:
    comm_df = pd.read_csv(data_path + comm)
    comm_nm = comm.split('.')[0]
    print(comm_nm + " : " + str(comm_df['count'].sum()))
```

```python
for comm in comm_pop_datasets:
    comm_df = pd.read_csv(data_path + comm)
    comm_nm = comm.split('.')[0]
    print(str(comm_df['count'].sum()))
```

```python
#List communities under 100 people
#For communities with under 100 people, how many towers are there?
comm_u100 = []

for comm in comm_pop_datasets:
        comm_df = pd.read_csv(data_path + comm)
        comm_nm = comm.split('.')[0]
        comm_size = comm_df['count'].sum()
        
        if comm_size < 100:
            comm_u100.append(comm_nm)
            print(comm_nm + ' : ' + str(len(comm_df['count']))+ " : " + str(comm_df['count'].sum()))
    
    #if comm_size < 100:
        #comm_u100.append(comm_nm)
        
#print(comm_u100)
len(comm_u100)
```

```python
comm_grid = gpd.read_file('../../data/working_files/community_wpopulation/community_0_pop.shp')

comm_grid.head()
```

```python
comm_grid['Comm_Pro'].sum()
```

```python
comm_grid_sort = comm_grid.sort_values(by=['Comm_Pro'], ascending=False)
comm_grid_sort.head()
```

```python
comm_columns = comm_grid_sort.columns.values
print(comm_columns)
```

```python
comm_comp = comm_grid_sort.drop(columns=['Join_Count', 'TARGET_FID', 'OBJECTID', 'IID', 'Total_Pop', 'Pro_Pop', 'Shape_Leng', 'Shape_Area', 'Join_Cou_1', 'TARGET_F_1', 'OBJECTID_1', 'community', 'count'])

comm_comp.head()
```

```python
comm_comp.insert(1, 'Comm_Per', 0)
comm_comp.head()
```

```python
total_comm_pop = comm_comp['Comm_Pro'].sum()
print(total_comm_pop)
```

```python
comm_comp['Comm_Per'] = comm_comp["Comm_Pro"].cumsum()
```

```python
comm_comp.insert(1, 'Comm_Perc', 0)

for row in range(len(comm_comp['Comm_Perc'])):
    comm_comp.loc[row, 'Comm_Perc'] = (comm_comp.loc[row, 'Comm_Per']/total_comm_pop) * 100

comm_comp.head()
```
